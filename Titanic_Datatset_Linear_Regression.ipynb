# Importing the Necessary Libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
print("Imported Libraries Successfully")
```

# Importing Our Titanic Dataset inbuilt in the Seaborn Library

```python
df = sns.load_dataset("titanic")
```

```python
df.head()
```

# Dropping the Unnecessary Columns in the Course of our Visualization

```python
df = df.drop(["deck", "embark_town", "alive", "who", "adult_male", "class"], axis = 1)
```

```python
df.head()
```

# Handling Missing Values in our Data

```python
df["age"] = df["age"].fillna (df["age"]. median())
df["embarked"] = df["embarked"].fillna (df["embarked"].mode ()[0])
```

# Mapping Numbers to 0 and 1

```python
# Mapping the Male to (0) and Female to (1)
df["sex"] = df["sex"].map({"male":0, "Female":1})
df["embarked"] = df["embarked"].map({"S":0, "C":1, "Q":2})
df.head(20)
```

# Plotting the Edited and Finished Titanic Dataset on a Graph

```python
# Survival Rate By Sex or Gender
sns.barplot(x = "sex", y = "survived", data = df)
plt.title("Survival Rate By Gender")
plt.show()
```

# Visulizing Survival Rate By Passenger Class

```python
# Survival Rate By Passenger Class
sns.barplot(x = "pclass", y = "survived", data = df)
plt.title("Survived Rate By Class")
plt.show()
```

# Using Histogram to Visualize the Age Distribution of Passengers on the Ship

```python
sns.histplot(df["age"], kde = True)
plt.title("Age Distribution of Passengers on the Ship")
plt.show()
```
