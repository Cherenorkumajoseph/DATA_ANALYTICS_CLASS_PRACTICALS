# Importing the Necessary Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
print("imported libraries successfully")
```

# Loadig the Students Data

```python
data = pd.read_csv("Students_data.csv")
data.head(50)
```

# Checking Null Values in the Data

```python
# Checking Null Values
data.info()
```

# Visualizing Our Data

```python
# Visualize the pass and fail Students
sns.countplot(x = "Pass", data = data)
plt.title("Class Distribution")
plt.show()
```

# Using Pair Plot to Visualize the Data

```python
# Pair Plot
sns.pairplot(data, hue = "Pass")
plt.show()
```

# Feature Encoding and Selection

```python
# Encode Categorical Variable
encoded_data = pd.get_dummies (data, columns = ["Social Activity"], drop_first = True)
encoded_data.head(50)
```

# Define Our Imput (X) Our Target y

```python
X = encoded_data.drop ("Pass", axis = 1)
y = encoded_data ["Pass"].map ({"Yes":1, "No":0})
```

# Splitting Our Data into Training and Testing

```python
# Splitting our data into Training and Testing
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split (X, y, test_size = 0.2, random_state = 1, stratify = y)
```

# Creating a Model

```python
from sklearn.tree import DecisionTreeClassifier

# Create the model
tree_model = DecisionTreeClassifier (criterion = "entropy", max_depth = 3, random_state = 1)
tree_model.fit(X_train, y_train)
```

# Visualizing the Decision Tree 

```python
# Visualize the Decision Tree
from sklearn.tree import plot_tree

plt.figure(figsize = (16,10))
plot_tree(tree_model, feature_names = X.columns, class_names = ["Fail", "Pass"], filled = True)
plt.title("Decision Tree - Students Pass Prediction")
plt.show()                                          
```

# Adding Confusion Matrix to our Visualization

```python
# Model Accuracy and Confusion Matrix
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
# Prediction
y_pred = tree_model.predict(X_test)

# Accuracy
print("Accuracy:", accuracy_score (y_test, y_pred))
print("Classification Report: \n", classification_report (y_test, y_pred))

# Confusion Matrix
cm = confusion_matrix (y_test, y_pred)
sns.heatmap(cm, annot = True, fmt = "d", cmap = "Blues")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```

# Saving the Model
