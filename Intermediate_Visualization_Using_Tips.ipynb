# Running Multiple Querries

```python
# Selecting Customers on Sunday that are Smokers
df[(df["day"] == "sun") & (df["smoker"] == "Yes")]
```

```python
# Selecting Customers on Friday that re smokers
df[(df["day"] == "Fri")  & (df["smoker"] == "Yes")]
```

```python
# Selecting Customers that that came on friday that are smokers and dinner time
df[(df["day"] == "Fri") & (df["smoker"] == "Yes") & (df["time"] == "Dinner")]
```

# Creating Another Column

```python
# Creating Another Column for Tip Percentage
df["tip percentage"] = (df["tip"] / df["total_bill"]) * 100
```

```python
df.head()
```

# Dropping or Deleting a Column

```python
# Dropping or Deleting the Tip Percentage Column
df.drop("tip percentage", axis = 1, inplace = True)
```

```python
df.head()
```

# Grouping the Data by Sex

```python
group = df.groupby("sex")
```

```python
group
```

```python
# Finally Grouping the Data
group.agg({"total_bill": "mean", "tip": "sum"})
```

# Plotting our data on a Line Chart

```python
# Line Chart
X = [22, 33, 10, 88, 60]
y = ["Mon", "Tue", "Wed", "Thur", "Fri"]
plt. plot (y, X, color = "Blue", label = "Sales Data")
plt.ylabel("Sales")
plt.xlabel("Days")
plt.title("Sales Trend")
plt.legend()
plt.show()
```

```python
plt.plot(df.index, df["total_bill"], color = "Red", linestyle = "dashed", label = "Total_Bill")
plt.xlabel("Index")
plt.ylabel("Total_Bills")
plt.title("Line Plot of Total Bills")
plt.legend()
plt.show()
```

# Plotting a Scatter Plot

```python
# Plotting on a Scatter Plot
plt.scatter(df["total_bill"], df["tip"], color = "Red", marker = "x")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.title("Scatter Plot: Total Bill Vs Tip")
plt.show()
```

# Plotting on  Bar Chart


```python
# Plotting on a Bar Chart
sns.barplot(x = "day", y = "tip", data = df, palette = "viridis")
plt.show()
```

# Plotting on a Pie Chart

```python
# Plotting on a Pie Chart
df["sex"].value_counts().plot(kind = "pie", colors = ["blue", "pink"], autopct = "% 1.1f %%")
plt.title("Gender Distribution of Customers")
plt.ylabel("")
plt.show()
```
