# Defining the Sigmoid Function

```python
# Sigmoid Function
def Sigmoid(Z):
    return 1/(1+np.exp(-Z))
```

# Testing the Sigmoid Function


```python
print(Sigmoid (1000))
```

```python
print(Sigmoid (1000000))
```

```python
print(Sigmoid (0))
```

```python
print(Sigmoid (-100000))
```

# Plotting the Sigmoid Numbers on a Graph

```python
# Plot
plt.figure(figsize = (10, 6))
plt.plot(Z, Sig, label = "Sigmoid (Z)", color = "Blue")
plt.title("Sigmoid Function with Classification Zone")
plt.xlabel("Z")
plt.ylabel("Sigmoid (Z)")
plt.grid("True")
plt.legend()
plt.show()
```

# Using the Sigmoid Numbers to Plot an Advanced Graph with Clear Illustration

```python
# Plot
plt.figure(figsize = (10, 6))
plt.plot (Z, Sig, label = "Sigmoid (Z)", color = "Blue")
plt.title("Sigmoid Function with Classification Zone")
plt.axhline(0.5, color = "Green", linestyle = "--", label = "Threshold = 0.5")
plt.axvline(0, color = "Red", linestyle = "--", label = "Z =0")
plt.text(4, 0.6, "Yes (1)", fontsize = 12, color = "Purple")
plt.xlabel("Z")
plt.ylabel("Sigmoid (Z)")
plt.grid("True")
plt.legend()
plt.show()
```
