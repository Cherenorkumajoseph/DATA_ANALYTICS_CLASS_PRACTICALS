# Importing the Necessary Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Importing our Dataset

```python
tips = sns.load_dataset("tips")
tips.head(50)
```

# Checking Non Values from our Imported Dataset

```python
tips.isnull().sum()
```

# Selecting data using .loc (when the meal happen on sun)

```python
# Get all records when the meal happened on Sunday
sun_meals = tips.loc[tips["day"] == "sun"]
sun_meals 
```

# Get Sunday Transactions with only Total_Bill and Tips Columns

```python
tips.loc[tips["day"] == "sun", ["total_bill", "tip", "day"]]
```

# Using .iloc to select first (10) rows in our Tips Dataset

```python
# First (10) rows
first_10_orders = tips.iloc[:10]
first_10_orders 
```

# Selecting first (10) rows and first (3) columns

```python
# First (10) rows and first (3) columns
subset_data = tips.iloc[:10, :3]
subset_data
```

# Selecting High Spenders in Weekends

```python
# High Spenders in weekends
high_spenders = tips.query("total_bill > 30 and smoker == 'No'")
high_spenders
```

# Selecting High Spenders in Weekends that are Smokers

```python
# High Spenders in Weekends that are smokers
high_spenders = tips.query("total_bill > 30 and smoker == 'Yes'")
high_spenders
```

# Selecting the High Tipers in Weekends

```python
# High Tipers in Weekends
high_tippers = tips.query("tip > 6 and day in ['sat', 'sun']")
high_tippers.head()
```

# .Merge() - Combining DataFrames

```python
# Creating Datase to Merge with out tips dataset
discount = pd.DataFrame({"day": ["sun", "sat", "Fri", "Thur"], "discount": [5, 10,7, 3]})
```

```python
# Using the .Merge to join our tips dataset with our new dataset pd.DataFrame
merged_tips = tips.merge(discount, on = "day", how = "left")
merged_tips 
```

```python
merged_tips.head(50)
```

# Using the Query command to Select for specific days seperately

```python
# Selecting for Sunday
merged_tips.query("day == 'sun'")
```

```python
# Selecting for Saturday
merged_tips.query("day == 'sat'")
```

```python
# Selecting for Friday
merged_tips.query("day =='Fri'")
```

```python
# Selecting for Thursday
merged_tips.query("day == 'Thur'")
```

# HEAT MAP USING OUR TIPS DATASET

```python
# Creating Heatmpa->Correlations
correlation_matrix = tips.corr(numeric_only = True)
```

# Creating the Heatmap Using the Created Correlation above

```python
# Creating the Heatmap
plt.figure(figsize = (8, 6))
sns.heatmap(correlation_matrix, annot = True, cmap = "coolwarm", linewidth = 0.5)
plt.title("Correlation Heatmap of Tips Dataset")
plt.show()
```

# Creating Advanced Correlation Heatmap

```python
# Advanced Correlation Heatmap
plt.figure(figsize = (10, 8))
sns.heatmap(
    correlation_matrix,
    annot = True, # Show Correlation values inside th cells.
    fmt = ".2f", # formats values to 2 decimal places.
    cmap = "viridis", # Change the colorscheme (Coolwarm, Blues, Reds, Magna, cividis).

    linewidths = 1, # Tickness.
    linecolor = "black",
    cbar = True,
    cbar_kws = {"shrink": 3, "aspect": 5},
    annot_kws = {"color": "black", "weight": "bold", "size": 12},
    square = True, # makes the cells perfect square.
    xticklabels = True,
    yticklabels = True
)
plt.title("Advaned Correlation Heatmap of Tips Dataset", fontsize = 16, fontweight = "bold")
plt.xticks(rotation = 45, fontsize = 12) # Rotates the x axis labels.
plt.yticks(fontsize = 12)
plt.show()
```
